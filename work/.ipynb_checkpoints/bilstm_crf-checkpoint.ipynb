```python
import pandas as pd
import glob
from tika import parser
def create_dict(text):
    parsed = parser.from_file(text)
    find_list = [["臨床診断", "検査目的"], ["臨床情報",  "部位"], ["読影所見", "読影診断"]]
    find_dict = {}
    for i in find_list:
        find_dict[i[0]] = parsed['content'][parsed['content'].find(i[0]) + len(i[0]):parsed['content'].find(i[1])].replace("\n", "")
    find_dict["読影診断"] = parsed['content'][parsed['content'].find("読影診断") + len("読影診断"):].replace("\n", "")
    return find_dict
def concat_DF_dict():
    some_df = pd.DataFrame([], columns=["臨床診断", "臨床情報", "読影所見", "読影診断"])
    dict_array = []
    for i in glob.glob('data/*.pdf'):
        some_dict = create_dict(i)
        dict_array.append(some_dict)
    some_df = pd.concat([some_df, pd.DataFrame.from_dict(dict_array)])
    return some_df

c_df = concat_DF_dict()
```

# LabelSpreading test

```python
import sklearn
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
from sklearn.semi_supervised import LabelPropagation
from sklearn.semi_supervised import LabelSpreading
from sklearn.datasets import make_circles
import random
import math
# 画像サイズを指定
plt.rcParams["figure.figsize"] = (14, 14)
```

```python
# データセット作成時のパラメータ
n_samples = 400  # 全データ数
factor = 0.4     # 内側の輪の直径を指定する係数 
label_rate = 0.1 # 学習に利用するラベル付きデータの割合
flip_rate = 0.1  # ラベルに加えるノイズ (ラベルを反転させる確率)
noise = 0.05     # 特徴量に加えるノイズ (ガウシアンノイズの強さ)
random_state = random.randint(1, 1000)  # 乱数のシードを指定して再現性を保つ
```

```python
# データセットを作成
X, y = make_circles(n_samples=n_samples, noise=noise, factor=factor,
                    shuffle=True, random_state=random_state)
```

```python
# 半教師あり学習に利用するラベル付きのデータの数
n_labeled = math.floor(n_samples * 0.5 * label_rate)

# ラベルの入れ物を作成
outer, inner = 0, 1  
labels = np.full(n_samples, -1.)

# 半教師あり学習に利用する "inner" ラベルを作成
labels[np.where(y==inner)[0][0:n_labeled]] = [int(random.uniform(0, 1) >= flip_rate) for x in range(n_labeled)]

# 半教師あり学習に利用する "outer" ラベルを作成
labels[np.where(y==outer)[0][0:n_labeled]] = [int(random.uniform(0, 1) <= flip_rate) for x in range(n_labeled)]
```

```python
plt.scatter(X[labels == outer, 0], X[labels == outer, 1], color='navy',
            marker='s', lw=0, label="outer labeled", s=10)
plt.scatter(X[labels == inner, 0], X[labels == inner, 1], color='c',
            marker='s', lw=0, label='inner labeled', s=10)
plt.scatter(X[labels == -1, 0], X[labels == -1, 1], color='darkorange',
            marker='.', label='unlabeled')
plt.legend(scatterpoints=1, shadow=False, 
           loc='upper left', bbox_to_anchor=(1.05, 1))
plt.title("Raw data (classes=outer, inner)")
plt.show()
```

```python
kernel = 'knn'
n_neighbors = 7

label_prop = LabelPropagation(kernel=kernel, max_iter=1000, n_jobs=-1, n_neighbors=n_neighbors)
label_prop.fit(X, labels)
```

```python
alpha = 0.2
n_neighbors = 7
kernel = 'knn'

label_spread = LabelSpreading(kernel=kernel, alpha=alpha, n_neighbors=n_neighbors,
                              max_iter=1000, n_jobs=-1)
label_spread.fit(X, labels)
```

```python
print("ラベル伝播法の精度 (accuracy) = %.2f" % (label_prop.score(X, y)))
print("ラベル拡散法の精度 (accuracy) = %.2f" % (label_spread.score(X, y)))
```

```python
# ラベル拡散法の学習結果をグラフ化
output_labels_spread = label_spread.transduction_
output_label_array = np.asarray(output_labels_spread)
outer_numbers = np.where(output_label_array == outer)[0]
inner_numbers = np.where(output_label_array == inner)[0]
plt.scatter(X[outer_numbers, 0], X[outer_numbers, 1], color='navy',
            marker='s', lw=0, s=10, label="outer learned")
plt.scatter(X[inner_numbers, 0], X[inner_numbers, 1], color='c',
            marker='s', lw=0, s=10, label="inner learned")
plt.legend(scatterpoints=1, shadow=False, 
           loc='upper left', bbox_to_anchor=(1.05, 1))
plt.title("Labels learned with Label Spreading (%s)" %(kernel))
plt.show()
```

```python
# ラベル伝播法の学習結果をグラフ化
output_labels_prop   = label_prop.transduction_
output_label_array = np.asarray(output_labels_prop)
outer_numbers = np.where(output_label_array == outer)[0]
inner_numbers = np.where(output_label_array == inner)[0]
plt.scatter(X[outer_numbers, 0], X[outer_numbers, 1], color='navy',
            marker='s', lw=0, s=10, label="outer learned")
plt.scatter(X[inner_numbers, 0], X[inner_numbers, 1], color='c',
            marker='s', lw=0, s=10, label="inner learned")
plt.legend(scatterpoints=1, shadow=False, 
           loc='upper left', bbox_to_anchor=(1.05, 1))
plt.title("Labels learned with Label Propagation (%s)" %(kernel))
plt.show()
```

# ginzaを使って半教師あり学習のデータ・セットを作成


```python
import spacy
from spacy.pipeline import EntityRuler 
from spacy import displacy

df_come = pd.read_csv("../Comejisyo/ComeJisyoUtf8-3/ComeJisyoUtf8-3.csv",  header=None)
df_come = df_come.dropna(how='all')

df_m = pd.read_excel("../data/MANBYO_20190704.xlsx")
df_m = df_m.dropna(how='all')

from tqdm import tqdm_notebook as tqdm 

class DependencyAnalysis:
    def __init__(self):
        self.nlp = spacy.load('ja_ginza')
        patterns = [{"label": "INFO", "pattern": "天気"},
           {"label": "INFO", "pattern": "気温"},
           {"label":"BY", "pattern":"頸椎"}]
        for i in tqdm(df_come[0]):
            patterns.append({"label":"Lesion", "pattern":i})
        for i in tqdm(df_m["出現形"]):
            patterns.append({"label":"Lesion", "pattern":i})
        ruler = EntityRuler(self.nlp)
        ruler.add_patterns(patterns)
        self.nlp.add_pipe(ruler) 
    def run(self, text):
        doc = self.nlp(text)

        displacy.render(doc, style="dep", jupyter=True, options={'distance': 90})
        displacy.render(doc, style="ent", jupyter=True,)
    def ents(self, text):
        doc = self.nlp(text)
        doc_list = []
        for ent in doc.ents:
            doc_list.append([
                ent.text, # テキスト
                ent.label_, # ラベル
                ent.start_char, # 開始位置
                ent.end_char # 終了位置
            ])
        return doc_list
        

dependency = DependencyAnalysis()
```

# 以下bi-LSTM CRF実装

```python
import torch
import torch.autograd as autograd
import torch.nn as nn
import torch.optim as optim

torch.manual_seed(1)
```

```python
# ベクトルの最大要素indexを返すargmax
def argmax(vec):
    # return the argmax as a python int
    _, idx = torch.max(vec, 1)
    return idx.item()
```

```python
# 系列データをインデックスのリストに変換する
def prepare_sequence(seq, to_ix):
    idxs = [to_ix[w] for w in seq]
    return torch.tensor(idxs, dtype=torch.long)
```

```python
def log_sum_exp(vec):
    max_score = vec[0, argmax(vec)]
    max_score_broadcast = max_score.view(1, -1).expand(1, vec.size()[1])
    # vec-max_scoreでmaxのタグは0、それ以外は負になる
    # exp(vec - max_score_broadcast)でmaxのタグは1、それ以外は1以下になる
    # よってvecのmaxが他と大きく離れていれば、そのsumのlogは0に近いので返り値はほとんどmax_scoreに等しい
    # 逆に、vecの中にmaxに近い値が含まれていれば、sumのlogは0より大きい値を持ち、返り値はmax_scoreを上回る
    # つまり、より確信度が高くタグを予想できればスコアは低く、曖昧な予想だとスコアが高くなる
    return max_score + \
        torch.log(torch.sum(torch.exp(vec - max_score_broadcast)))
```

```python
class BiLSTM_CRF(nn.Module):
    def __init__(self, vocab_size, tag_to_ix, embedding_dim, hidden_dim):
        super(BiLSTM_CRF, self).__init__()
        self.embedding_dim = embedding_dim
        self.hidden_dim = hidden_dim
        self.vocab_size = vocab_size
        self.tag_to_ix = tag_to_ix
        self.tagset_size = len(tag_to_ix)

        # LSTM part
        self.word_embeds = nn.Embedding(vocab_size, embedding_dim)
        # bi-LSTMするために bidirectional=True にしている
        self.lstm = nn.LSTM(embedding_dim, hidden_dim // 2,
                            num_layers=1, bidirectional=True)
        self.hidden2tag = nn.Linear(hidden_dim, self.tagset_size)

        # CRF part
        # あるNEタグから別のNEタグへ遷移するスコアの行列（論文内での行列A)
        # k×kの正方行列（k: タグの数）
        # i行j列の成分は、jのタグからiのタグへの遷移スコア
        self.transitions = nn.Parameter(
            torch.randn(self.tagset_size, self.tagset_size))
        # STARTタグへの遷移、STOPタグからの遷移は起こりえないので-10000と設定
        self.transitions.data[tag_to_ix[START_TAG], :] = -10000
        self.transitions.data[:, tag_to_ix[STOP_TAG]] = -10000

        self.hidden = self.init_hidden()

    # 隠れ状態の初期化関数
    def init_hidden(self):
        return (torch.randn(2, 1, self.hidden_dim // 2),
                torch.randn(2, 1, self.hidden_dim // 2))    

    # featsはLSTMの出力、n×k行列 これは損失関数の計算に使われます。
    def _forward_alg(self, feats):
        # 大きさが1×k、値は全て-10000の行列を用意
        init_alphas = torch.full((1, self.tagset_size), -10000.)
        init_alphas[0][self.tag_to_ix[START_TAG]] = 0.

        forward_var = init_alphas

        for feat in feats:
            alphas_t = []
            for next_tag in range(self.tagset_size):
                # featの中のnext_tagに該当する値を1×kの行列に引き延ばす（値は同じ）
                # feat[next_tag] = tensor(0.3447)
                # emit_score = tensor([[ 0.3447,  0.3447,  0.3447,  0.3447,  0.3447]])
                emit_score = feat[next_tag].view(
                    1, -1).expand(1, self.tagset_size)
                
                # タグからタグへの遷移行列（論文中のA）から注目しているnext_tagへの遷移スコアを取得
                # trans_scoreのi番目はiのタグからnext_tagへの遷移スコア
                trans_score = self.transitions[next_tag].view(1, -1)
                
                # next_tag_varのi番目はiのタグからnext_tagへ遷移するスコア
                # featのfor文の2週目以降（2つ目の単語以降）はそれまでの単語が算出したforwardが効いてくる
                next_tag_var = forward_var + trans_score + emit_score
                # next_tagへ遷移するスコアをlog_sum_expの合計で計算
                alphas_t.append(log_sum_exp(next_tag_var).view(1))
            # それぞれのタグへの遷移スコア
            # 2つ目の単語以降はそれまでの単語がどのタグに遷移しやすいかのスコアが効いてくる
            forward_var = torch.cat(alphas_t).view(1, -1)
        # STOPへの遷移スコアを加える
        terminal_var = forward_var + self.transitions[self.tag_to_ix[STOP_TAG]]
        # 返り値alphaはそれぞれの単語ごとのタグの遷移に関して、確信度が高く予想できれば小さい値、
        # 逆に曖昧な予想が多ければ大きい値になる
        alpha = log_sum_exp(terminal_var)
        return alpha
    def _get_lstm_features(self, sentence):
        self.hidden = self.init_hidden()
        embeds = self.word_embeds(sentence).view(len(sentence), 1, -1)
        lstm_out, self.hidden = self.lstm(embeds, self.hidden)
        lstm_out = lstm_out.view(len(sentence), self.hidden_dim)
        lstm_feats = self.hidden2tag(lstm_out)
        return lstm_feats
    # s(X, y)の計算
    def _score_sentence(self, feats, tags):
        score = torch.zeros(1)
        tags = torch.cat([torch.tensor([self.tag_to_ix[START_TAG]], dtype=torch.long), tags])
        for i, feat in enumerate(feats):
            score = score + \
                self.transitions[tags[i + 1], tags[i]] + feat[tags[i + 1]]
        score = score + self.transitions[self.tag_to_ix[STOP_TAG], tags[-1]]
        return score
    def _viterbi_decode(self, feats):
        backpointers = []

        # タグからタグへの遷移スコア
        init_vvars = torch.full((1, self.tagset_size), -10000.) # tensor([[-10000., -10000., -10000., -10000., -10000.]])
        # STARTタグへの遷移はあり得ないので0に設定
        init_vvars[0][self.tag_to_ix[START_TAG]] = 0 # tensor([[-10000., -10000., -10000.,      0., -10000.]])

        # LSTMの出力一つ一つに関して遷移スコアを計算する
        forward_var = init_vvars
        for feat in feats:
            bptrs_t = []
            viterbivars_t = []

            # 
            for next_tag in range(self.tagset_size):
                next_tag_var = forward_var + self.transitions[next_tag]
                best_tag_id = argmax(next_tag_var)
                bptrs_t.append(best_tag_id)
                viterbivars_t.append(next_tag_var[0][best_tag_id].view(1))
            forward_var = (torch.cat(viterbivars_t) + feat).view(1, -1)
            backpointers.append(bptrs_t)

        terminal_var = forward_var + self.transitions[self.tag_to_ix[STOP_TAG]]
        best_tag_id = argmax(terminal_var)
        path_score = terminal_var[0][best_tag_id]

        best_path = [best_tag_id]
        for bptrs_t in reversed(backpointers):
            best_tag_id = bptrs_t[best_tag_id]
            best_path.append(best_tag_id)
        start = best_path.pop()
        assert start == self.tag_to_ix[START_TAG]
        best_path.reverse()
        return path_score, best_path
    def neg_log_likelihood(self, sentence, tags):
        feats = self._get_lstm_features(sentence)
        forward_score = self._forward_alg(feats)
        gold_score = self._score_sentence(feats, tags)
        return forward_score - gold_score
    def forward(self, sentence):
        lstm_feats = self._get_lstm_features(sentence)
        score, tag_seq = self._viterbi_decode(lstm_feats)
        return score, tag_seq
```

```python
# IBOタグに加えてSTART、STOPタグを追加、よってk=5
START_TAG = "<START>"
STOP_TAG = "<STOP>"
EMBEDDING_DIM = 5
HIDDEN_DIM = 4

# 学習データ、単語のリストとNEタグのリスト
training_data = [(
    "the wall street journal reported today that apple corporation made money".split(),
    "B I I I O O O B I O O".split()
), (
    "georgia tech is a university in georgia".split(),
    "B I O O O O B".split()
)]

# 単語にインデックスを振る
word_to_ix = {}
for sentence, tags in training_data:
    for word in sentence:
        if word not in word_to_ix:
            word_to_ix[word] = len(word_to_ix)

# タグにインデックスを振る
tag_to_ix = {"B": 0, "I": 1, "O": 2, START_TAG: 3, STOP_TAG: 4}

# モデルと最適化アルゴリズムのインスタンス生成
# 損失関数はBiLSTM_CRF()内で設定したものを使用
model = BiLSTM_CRF(len(word_to_ix), tag_to_ix, EMBEDDING_DIM, HIDDEN_DIM)
optimizer = optim.SGD(model.parameters(), lr=0.01, weight_decay=1e-4)

# 学習前のスコアとタグをチェック
with torch.no_grad():
    precheck_sent = prepare_sequence(training_data[0][0], word_to_ix)
    precheck_tags = torch.tensor([tag_to_ix[t] for t in training_data[0][1]], dtype=torch.long)
    print(model(precheck_sent)) 
    # タグの予測が間違っている
    # (tensor(2.6907), [1, 2, 2, 2, 2, 2, 2, 2, 2, 2, 1])

# 学習開始
for epoch in range(
        300):  # 300エポックは実務では小さすぎる、ただのお試し
    for sentence, tags in training_data:
        # Step 1. 勾配の初期化
        model.zero_grad()

        # Step 2. 入力データをインデックスの形に変換
        sentence_in = prepare_sequence(sentence, word_to_ix)
        targets = torch.tensor([tag_to_ix[t] for t in tags], dtype=torch.long)

        # Step 3. フォワードを走らせて損失関数を計測
        loss = model.neg_log_likelihood(sentence_in, targets)

        # Step 4. バックワード、パラメータの更新
        loss.backward()
        optimizer.step()

# 学習後のスコアと予測タグをチェック
with torch.no_grad():
    precheck_sent = prepare_sequence(training_data[0][0], word_to_ix)
    print(model(precheck_sent))
    # スコアも上がっているし、正しくタグを予測できている
    # (tensor(20.4906), [0, 1, 1, 1, 2, 2, 2, 0, 1, 2, 2])
```
